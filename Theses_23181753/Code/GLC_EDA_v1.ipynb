```python
for col in ["Landslide_date", "start_date_30d", "start_date_60d"]:
    Rainfalldatachirp[col] = pd.to_datetime(Rainfalldatachirp[col])
```

```python
file_pattern = "chirps-v2.0.*.days_p05"
```

```python
import numpy as np
```

```python
def getprecipitatindata(df, lat,lon,startdt,enddt):
    lat_idx = np.abs(df['latitude'] - lat).argmin()
    lon_idx = np.abs(df['longitude'] - lon).argmin()
    precip_values = df.sel(time=slice(startdt, enddt)).precip
    cumulative_precip = precip_values[:, lat_idx, lon_idx].sum().item()
    return cumulative_precip
```

```python
Rainfalldatachirp.head(5)
```

```python
results = []
for _, row in Rainfalldatachirp.iterrows():
    Landslide_date = row['Landslide_date']
    year = Landslide_date.year


    file_path = f"CHIRPSfile/chirps-v2.0.{year}.days_p05.nc"
    print("filepath1",file_path)
    try:
        with xr.open_dataset(file_path, engine="netcdf4") as rdata:
            print("filepath2",file_path)
            precip_30d = getprecipitatindata(rdata, row['latitude'], row['longitude'], row['start_date_30d'], row['Landslide_date'])
            precip_60d = getprecipitatindata(rdata, row['latitude'], row['longitude'], row['start_date_60d'], row['Landslide_date'])
            #precip_7d = getprecipitatindata(rdata, row['latitude'], row['longitude'], row['start_date_7d'], row['Landslide_date'])

            results.append({
                "Landslide_date": Landslide_date,
                "latitude": row['latitude'],
                "longitude": row['longitude'],
                "precip_30d": precip_30d,
                "precip_60d": precip_60d,
                #"precip_7d": precip_7d,
                #"is_rainfall_induced": row['is_rainfall_induced']
                "landslide_trigger" : row['landslide_trigger']
            })
    except FileNotFoundError:
        print(f"NetCDF file for year {year} not found.")
    except Exception as e:
        print(f"Error processing landslide for date {Landslide_date}: {e}")

```

```python
precipitation_data = pd.DataFrame(results)
```
